```python
import matplotlib.pyplot as plt
```

## 折线图 plot

```python
y_label=["11point{}minutes".format(i) for i in range(10)]

plt.figure(figsize=(5,5))#figsize=(长，宽) 画布大小
plt.plot(range(10),range(10))

plt.xticks(range(10)) #修改刻度


plt.yticks(range(10),y_label)

# plt.xlabel('') 横坐标描述信息
# plt.title('') 标题

# plt.legend() 显示图例
plt.show()
# plt.savefig('a.png') 保存文件
```
